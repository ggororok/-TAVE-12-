```python
df = pd.read_csv("C:/Users/pc/Desktop/민지/동아리/프로젝트/리뷰정보_4,5점과50개이상.csv", encoding='CP949')
df.drop(["Unnamed: 0"], axis=1, inplace=True)
print(df.shape)
df.head(5)
```

```python
review = df[['제품명','리뷰 내용']]
review.head(5)
```

```python
#특수문자 및 띄어쓰기 제거
import re

def remove_special_characters(text):
    cleaned_text = re.sub(r'[^가-힣0-9]', ' ', text) #한글이 아닌 모든 문자 제거
    cleaned_text = ' '.join(cleaned_text.split())
    return cleaned_text

review['리뷰내용_new'] = review['리뷰 내용'].apply(remove_special_characters)

review.head(5)
```
